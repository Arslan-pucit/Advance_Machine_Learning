```python
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
anscomb_df = sns.load_dataset('anscombe')
anscomb_df
```

```python
x1 = np.array(anscomb_df[anscomb_df.dataset == 'I']['x'])
y1 = np.array(anscomb_df[anscomb_df.dataset == 'I']['y'])
x2 = np.array(anscomb_df[anscomb_df.dataset == 'II']['x'])
y2 = np.array(anscomb_df[anscomb_df.dataset == 'II']['y'])
x3 = np.array(anscomb_df[anscomb_df.dataset == 'III']['x'])
y3 = np.array(anscomb_df[anscomb_df.dataset == 'III']['y'])
x4 = np.array(anscomb_df[anscomb_df.dataset == 'IV']['x'])
y4 = np.array(anscomb_df[anscomb_df.dataset == 'IV']['y'])
n = len(anscomb_df[anscomb_df.dataset == 'I']['x'])
```

```python
x = [x1, x2, x3, x4]
y = [y1, y2, y3, y4]
beta_0  = []
beta_1  = []
```

# (i) Using Batch/Vanilla Gradient Descent (BGD): 
### Apply on four dataset in a single loop

```python
# initial points or the initial guess 
beta0 = 0
beta1 = 1
# Defining step size
epochs = 10000
alpha = 0.01
for j in range(len(x)):
    for i in range(epochs):
        y_hat = beta0 + beta1*x[j]
        
        # finding the slope of error fn. at beta0 and beta1
        d_beta0 =  -1/n * sum((y[j] - y_hat))
        d_beta1 =  -1/n * sum(x[j] * (y[j] - y_hat))

        # magic lines of this code when we updating the weights
        beta0 = beta0 - alpha*d_beta0
        beta1 = beta1 - alpha*d_beta1
    beta_0.append(beta0)
    beta_1.append(beta1)    
```

```python

for i in range(3):
    fig,ax = plt.subplots()
    X = np.linspace(4,14,20)
    Y = beta_0[i] + beta_1[i] * X
    ax.scatter(x[i],y[i])
    ax.plot(X, Y)
    plt.show()
fig,ax = plt.subplots()
X = np.linspace(8,18,20)
Y = beta_0[3] + beta_1[3] * X
ax.scatter(x[3],y[3])
ax.plot(X, Y)
plt.show()
```

# (ii) Using Stochastic Gradient Descent (SGD):
### (a) FIrst data set

```python
# initial points or the initial guess 
beta0 = 0
beta1 = 1
# Defining step size
epochs = 100000
alpha = 0.01
for i in range(epochs):
    index = np.random.randint(0,len(x1))
    
    y_hat = beta0 + beta1*x1[index]


    # finding the slope of error fn. at beta0 and beta1
    d_beta0 =  -1/1 * (y1[index] - y_hat)
    d_beta1 =  -1/1 * (x1[index] * (y1[index] - y_hat))

    # magic lines of this code when we updating the weights
    beta0 = beta0 - alpha*d_beta0
    beta1 = beta1 - alpha*d_beta1
```

```python
fig,ax = plt.subplots()
X = np.linspace(4,14,20)
Y = beta0 + beta1 * X
ax.scatter(x1,y1)
ax.plot(X, Y)
plt.show()
```
